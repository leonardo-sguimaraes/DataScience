## Reduce 

```python
# importando a função reduce do módulo functools
from functools import reduce
```

```python
lista =  [47,20,81,13]
```

```python
lista
```

```python
def soma(a,b):
    x = a +b
    return x
```

```python
reduce(soma, lista)
```
