## Dataframes

###### %Dataframes representam uma estrutura tabilar semelhante a estrutura de uma planilha do Excel, contendo uma coleção de colunas em que cada uma pode ser um diferente tipo de valor (número, string, etc...) Os Dataframes possuem index e linhas e esta estrutura é muito semehante a um dataframe em R. Os dados de um dataframe são armazenados e um ou mais blocos bidimensionais, ai invés de listas, dicionários ou alguma outra estrutura de array.

```python
from pandas import DataFrame
```

```python
data = {'Estado': ['Santa Catarina', 'Paraná', 'Goiás', 'Bahia', 'Minas Gerais'], 
       'Ano': [2002, 2003, 2004, 2005, 2006],
       'População': [1.5,1.7,3.6,2.4,2.9]}
```

## Time Series

```python
# Criando um range de datas com frequência de segundos

rng = pd.date_range('1/1/2016', periods=50, freq='S')
```

```python
ts =  pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
```

```python
ts
```

```python
# Criando um range de datas com Frequência de meses

rng = pd.date_range('1/1/2016', periods=5, freq='M')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

## Plotting

```python
import matplotlib.pyplot  as plt
from matplotlib import style
%matplotlib inline
```

```python
# Time Series Plot

ts = pd.Series(np.random.randn(500), index=pd.date_range('1/1/2016', periods=500))
ts = ts.cumsum()
ts.plot()
```

```python
#DataFrame Plot 

df = pd.DataFrame(np.random.randn(500,4), index=ts.index, columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
plt.figure(); df.plot(); plt.legend(loc = 'best')
```
